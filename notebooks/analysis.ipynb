```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import mysql.connector
```

```python
# Establish Database Connection (Here, with your MySQL Server)
connection = mysql.connector.connect(
    host="username",        # your DB host
    user="root",            # your username
    password="password",    # your password
    database="customers"    # database name
)
```

```python
# Monthly Revenue Trend Analysis

# SQL Query:

# Extract month from purchase timestamp
# Aggregate total revenue per month
# Sort results chronologically

query = """
SELECT 
    DATE_FORMAT(o.order_purchase_timestamp, '%Y-%m-01') AS month,
    SUM(p.payment_value) AS revenue
FROM orders o
JOIN payments p 
    ON o.order_id = p.order_id
GROUP BY month
ORDER BY month;
"""

# Load data into pandas DataFrame
df = pd.read_sql(query, connection)

# Data Cleaning & Preparation
df['month'] = pd.to_datetime(df['month'])

# Ensure data is sorted by time
df = df.sort_values('month')

# Visualization

df.plot(x='month', y='revenue')
plt.title("Monthly Revenue Trend")

# Display the plot
plt.show()

# Close Database Connection
connection.close()
```
